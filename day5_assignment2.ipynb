```python
list1 = [10, 20, 40, 60, 70, 80]
list2 = [5, 15, 25, 35, 45, 60]
```

```python
list3 = list1+list2
list3
```

```python
for i in range(0, len(list3)):    
    for j in range(i+1, len(list3)):    
        if(list3[i] > list3[j]):    
            temp = list3[i];    
            list3[i] = list3[j];    
            list3[j] = temp;    
            
print(list3)
```
